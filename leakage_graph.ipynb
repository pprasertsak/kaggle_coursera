# Graph Visualization of the Programming Assignment Week 2

```python
import random
import numpy as np

import matplotlib.pyplot as plt

C = 9 # the number of classes
I = 100 # the number of instances in each class

GRID = (3, 3)

grouped_samples = [0.1*np.random.randn(I, 2) + [x+1, y+1] for x in range(GRID[0]) for y in range(GRID[1])]
samples = np.vstack(grouped_samples)

E0 = 250 # the number of pairs that contain different classes
E1 = 250 # the number of pairs that contain the same class

random_E0 = np.random.choice(C*I, E0, replace=False)
E0_pairs = [(samples[i], random.choice(random.choice(grouped_samples[:i//I]+grouped_samples[i//I+1:]))) for i in random_E0]

random_E1 = np.random.choice(C*I, E1, replace=False)
E1_pairs = [(samples[i], random.choice(grouped_samples[i//I])) for i in random_E1]

colors = np.repeat(np.random.rand(C), I)

plt.figure(figsize=(9, 9))
plt.scatter(samples[:, 0], samples[:, 1], c=colors, s=10)

for pair in E0_pairs:
    plt.plot([pair[0][0], pair[1][0]], [pair[0][1], pair[1][1]], c='tab:gray', linewidth=0.5)
    
for pair in E1_pairs:
    plt.plot([pair[0][0], pair[1][0]], [pair[0][1], pair[1][1]], c='r', linewidth=1)

plt.show()
```

This is a graph of C classes, each of which has I instances. There are E0 connections between the classes and E1 connections within the classes, which are the gray lines and the red lines respectively. The classes are clustered along the grids for easy visualization. Note that one can vary these parameters for better modeling the real problems.
